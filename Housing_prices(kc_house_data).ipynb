```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
hp=pd.read_csv(r"C:\Users\BHAVIKKSVSS\Desktop\kc_house_data.csv")
hp.head()
```

```python
hp.info()
```

```python
hp=hp.iloc[:,1:]
hp.head()
```

```python
enc=LabelEncoder()
hp.iloc[:,14]=enc.fit_transform(hp.iloc[:,14])
```

```python
hp.head()
```

```python
feature_cols = ['bedrooms','bathrooms','sqft_living','sqft_lot','condition','floors','yr_built','sqft_above','lat','sqft_living15','sqft_lot15'] 
x = hp[feature_cols] # predictor
y = hp.price # response
```

```python
x.head()
```

```python
y
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.1) 
```

```python
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
linreg.score(x_test,y_test)
```
